###**Data Wragling**

```python
def group_age(x):
    if x < 20:
        return "Under 20"
    elif x >= 20 and x < 30:
        return "20-29"
    elif x > 30 and x < 40:
        return "30-39"
    elif x >= 40 and x < 50:
        return "40-49"
    else:
        return "More than 50"

def group_nationality(x):
    if x == "US":
        return "US"
    elif x == "Mexico":
        return "Mexico"
    elif x == "Caribean":
        return "Caribean"
    elif x == "Germany":
        return "Germany"
    elif x == "UK":
        return "UK"
    else:
        return "Others"
```

```python
for df in [train, test]:
    # Age
    df["age"] = df["age"].apply(group_age)
    df["age"].astype("object")
    
    # Work type
    df["work_type"].replace("?", "Private", inplace=True)
    df["work_type"].replace(("Local-gov", "State-gov", "Federal-gov"), ("Government", "Government", "Government"), inplace=True)
    df["work_type"].replace(("Self-emp-not-inc", "Never-worked"), ("Without-pay", "Without-pay"), inplace=True)
    
    # Job
    train["job"].replace("?", "Other-service", inplace=True)
    
    # Education
    train["education"] = train["education"].replace(("9th", "10th", "11th", "12th"),("High-school","High-school", "High-school", "High-school"))
    train["education"] = train["education"].replace(("5th-6th", "7th-8th"),("Middle-school", "Middle-school"))
    train["education"] = train["education"].replace(("1st-4th"), "Elementary-school")
    train["education"].replace(('Preschool', 'High-school', 'Middle-school', 'Elementary-school'), ('Popular-program', 'Popular-program', 'Popular-program', 'Popular-program'), inplace=True)
    train["education"].replace(("Assoc-voc", "Assoc-acdm", "Some-college"), ("Assoc/College","Assoc/College", "Assoc/College"), inplace=True)
    train["education"].replace(("Prof-school", "Doctorate"), ("PhD", "PhD"), inplace=True)
    
    # Nationality
    train["nationality"] = train["nationality"].replace(("?", "Puerto-Rico"), ("US", "US"))
    train["nationality"] = train["nationality"].replace(("England", "Scotland", "Ireland"), ("UK", "UK", "UK"))
    train["nationality"] = train["nationality"].replace(("Jamaica", "Dominican-Republic", "Trinadad&Tobago"), ("Caribean", "Caribean", "Caribean"))
    train["nationality"] = train["nationality"].apply(group_nationality)
```

```python
def encode(frame, feature):
    # Encode categorical variables
    ordering = pd.DataFrame()
    ordering['val'] = frame[feature].unique()
    ordering.index = ordering.val

    ordering['ordering'] = range(1, ordering.shape[0]+1)
    ordering = ordering['ordering'].to_dict()
    
    for cat, o in ordering.items():
        frame.loc[frame[feature] == cat, feature+'_E'] = o

def Scaler(X_train, X_val, X_test, index_scaler=0):
    # Feature scaling 
    list_scaler = [MinMaxScaler(), StandardScaler(),  Normalizer(), QuantileTransformer()]
    
    scaler = list_scaler[index_scaler]
    X_train = scaler.fit_transform(X_train)
    X_val = scaler.transform(X_val)  
    X_test = scaler.transform(X_test)  
    return X_train, X_val, X_test
```

```python
numerical_feats = train.dtypes[train.dtypes != "object"].index
print("Number of Numerical features: ", len(numerical_feats))

categorical_feats = train.dtypes[train.dtypes == "object"].index
print("Number of Categorical features: ", len(categorical_feats))
```

```python
cate_encoded = []
for q in categorical_feats:  
    encode(train, q)
    encode(test, q)
    cate_encoded.append(q+'_E')
print(cate_encoded)
```

```python
train_feats = []
train_feats.extend(cate_encoded)
train_feats.extend(numerical_feats)
train_feats = [cols for cols in train_feats if cols != "ID" and cols != "target_income"]
train_feats
```

```python
test_feats = []
test_feats.extend(cate_encoded)
test_feats.extend(numerical_feats)
test_feats = [cols for cols in test_feats if cols != "ID" and cols != "target_income"]
test_feats
```

```python
X = train[train_feats]
X_test = test[test_feats]
y = train['target_income']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3)
```

```python
X_train, X_val, X_test = Scaler(X_train, X_val, X_test, 0)
X_train = X_train.astype('float32')
X_val = X_val.astype('float32')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
from tensorflow.keras import utils
from tensorflow.keras import optimizers
import os

```

```python
checkpoint_path = "/content/save_model/model_nn.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    monitor='val_accuracy',
    save_weights_only=True,
    mode='max',
    save_best_only=True
)

# Set random seed for reproducability
tf.random.set_seed(1234)
initializer = tf.keras.initializers.RandomNormal()
```

```python
model = Sequential([
    Dense(32, activation='relu', input_dim = len(train_feats), kernel_initializer=initializer),
    BatchNormalization(),
    Dense(32, activation='relu', kernel_initializer=initializer),
    BatchNormalization(),
    Dense(32, activation='relu', kernel_initializer=initializer),
    BatchNormalization(),
    Dense(32, activation='relu', kernel_initializer=initializer),
    BatchNormalization(),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
#hyper-parameters for optimizer
learning_rate = 1e-4
opt = optimizers.Adam(learning_rate=learning_rate)

model.compile(loss='categorical_crossentropy', 
              optimizer=opt, metrics=['accuracy'])

epochs = 50
history = model.fit(X_train, y_train, epochs=epochs, 
                    validation_data=(X_val, y_val), 
                    callbacks=[model_checkpoint_callback])
```

```python
os.listdir(checkpoint_dir)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
model.load_weights(checkpoint_path)
test_predict = model.predict(X_val)
test_predict = np.around(test_predict)

print("Accuracy:", accuracy_score(y_val, test_predict))
```
