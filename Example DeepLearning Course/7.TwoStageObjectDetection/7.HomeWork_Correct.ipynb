#6) Xây dựng hàm tính toán IoU giữa hai bounding box.

```python
def iou(boxA, boxB):
  '''
    boxA: x_a, y_a, w_a, h_a
    boxB: x_b, y_b, w_b, h_b
  '''
  boxA=[boxA[0],boxA[1],boxA[0]+boxA[2],boxA[1]+boxA[3]]
  boxB=[boxB[0],boxB[1],boxB[0]+boxB[2],boxB[1]+boxB[3]]
  # Toạ độ hình chữ nhật tương ứng phần giao nhau
  xA = max(boxA[0], boxB[0])
  yA = max(boxA[1], boxB[1])
  xB = min(boxA[2], boxB[2])
  yB = min(boxA[3], boxB[3])
  
  # Tính diện tích phần giao nhau
  interArea = (xB - xA + 1) * (yB - yA + 1)
  
  # Diện tích của predicted và ground-truth bounding box
  boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
  boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
  
  # Tính IoU = diện tích phần giao nhau chia diện tích phần tổng hợp
  # Diện tích phần hợp = tổng diện tích trừ diện tích phần giao
  iou = interArea / float(boxAArea + boxBArea - interArea)

  return iou

iou([10, 20, 50, 50], [15, 25, 50, 50])
```

#7) Tính AP 

```python
list_predict=[[10, 10, 50, 20], [15, 15, 55, 25], [50, 20, 20, 25], [20, 50, 20, 50], [30, 20, 10, 10]]
ground_truth=[[10, 10, 45, 25], [45, 20, 20, 25], [30, 25, 10, 10]]
```

```python
import matplotlib.pyplot as plt

count=0
pos=0

prec=[]
rec=[]
for box in list_predict:
  count+=1
  for gth in ground_truth:
    iou_=iou(box,gth)
    if(iou_>=0.5):
      pos+=1
      break
  precision=pos/count
  recall=pos/len(ground_truth)
  prec.append(precision)
  rec.append(recall)
  print(precision)
  print(recall)
  print("___")

plt.xlabel('recall')
plt.ylabel('precision')
plt.plot(rec, prec)
plt.title('Precision Recal Curve')  
```

```python
def voc_ap(rec, prec):
    rec.insert(0, 0.0) # insert 0.0 at indice 0 of list 
    rec.append(1.0) # insert 1.0 at the end of list
    mrec = rec[:]
    prec.insert(0, 0.0) # insert 0.0 at indice 0 of list
    prec.append(0.0) # insert 0.0 at the end of list
    mpre = prec[:]

    # mpred is modify precision value in which
    # in a sort interval of recall select the higher value of precision 
    for i in range(len(mpre) - 2, 0, -1): # range(start, end, step)
        mpre[i] = max(mpre[i], mpre[i+1])
    i_list = []
    # mrec is modified recall from low to high.
    for i in range(1, len(mrec), 1):
        if mrec[i] != mrec[i-1]:
            i_list.append(i) 
    ap = 0.0
    for i in i_list:
        ap += ((mrec[i]-mrec[i-1])*mpre[i])
    return ap, mrec, mpre
```

# 9) Biến đổi COCO format

```python
import numpy as np
import json
import pandas as pd

path = 'train.csv' # the path to the CSV file
save_json_path = 'coco.json'

data = pd.read_csv(path)

images = []
categories = []
annotations = []

category = {}
category["supercategory"] = 'none'
category["id"] = 0
category["name"] = 'None'
categories.append(category)

data['fileid'] = data['image_id'].astype('category').cat.codes
data['categoryid']= data['class_id']
data['annid'] = data.index
def image(row):
    image = {}
    image["height"] = 0
    image["width"] = 0
    image["id"] = row.fileid
    image["file_name"] = row.image_id
    return image

def category(row):
    category = {}
    category["supercategory"] = 'None'
    category["id"] = row.categoryid+1
    category["name"] = row[2]
    return category

def annotation(row):
    annotation = {}
    area = (row.x_max -row.x_min)*(row.y_max - row.y_min)
    annotation["segmentation"] = []
    annotation["iscrowd"] = 0
    annotation["area"] = area
    annotation["image_id"] = row.fileid

    annotation["bbox"] = [row.x_min, row.y_min, row.x_max -row.x_min,row.y_max-row.y_min ]

    annotation["category_id"] = row.categoryid+1
    annotation["id"] = row.annid
    return annotation

for row in data.itertuples():
    annotations.append(annotation(row))

imagedf = data.drop_duplicates(subset=['fileid']).sort_values(by='fileid')
for row in imagedf.itertuples():
    images.append(image(row))

catdf = data.drop_duplicates(subset=['categoryid']).sort_values(by='categoryid')
for row in catdf.itertuples():
    categories.append(category(row))

data_coco = {}
data_coco["images"] = images
data_coco["categories"] = categories
data_coco["annotations"] = annotations
json.dump(data_coco, open(save_json_path, "w"), indent=4)
```
