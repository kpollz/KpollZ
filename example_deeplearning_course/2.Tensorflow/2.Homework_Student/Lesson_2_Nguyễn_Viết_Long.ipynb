# 9. Data preprocesing and train/test split

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/lesson 2/train.csv").dropna()
df.sample(10)
```

```python
df = df.drop("ID", axis = 1)
```

```python
df.info()
```

```python
# Observe the outliers by histogram 
df['age'].hist(bins=20)
```

```python
df['hrs_per_week'].hist(bins=10)
```

```python
df['final_weight'].hist()
```

```python
# Remove outliers from dataframe
df = df[df['age']<75]
df = df[df['hrs_per_week']>12]
df = df[df['hrs_per_week']<65]
df = df[df['final_weight']<600000]
```

```python
df.shape
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
# Label Encoding
le = LabelEncoder()
col = ['work_type', 'education', 'marital_state', 'job', 'ethnicity', 'status', 'sex', 'nationality']
for i in col:
    df[i] =  le.fit_transform(df[i])
df
```

```python
# Rescaling the data 
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
scaler = MinMaxScaler(feature_range=(0, 1)) 
x = scaler.fit_transform(x) 
x
```

```python
from sklearn.model_selection import train_test_split
# Split data 70/30 into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, 
                                                    stratify = y, # assign equal proportion of target label in train/test 
                                                    random_state = 0) # keep train/test split the same if run again. 
```

#10. Train model and evaluate 

```python
from tensorflow.keras import models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils
```

```python
# Set data types for float features
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

# Set data types for categorical labels
y_train = utils.to_categorical(y_train)
y_test = utils.to_categorical(y_test)
print('Ready...')
```

```python
y_test[500]
```

```python
x_train.shape
```

```python
# Build the classifier model
model = Sequential()
model.add(Dense(14, input_shape = (14,), activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='softmax'))
print(model.summary())
```

```python
# Train model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

num_epochs = 100
history = model.fit(x_train, y_train, epochs=num_epochs, batch_size=200, validation_data=(x_test, y_test))
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

epoch_nums = range(1,num_epochs+1)
training_loss = history.history["loss"]
validation_loss = history.history["val_loss"]
plt.plot(epoch_nums, training_loss)
plt.plot(epoch_nums, validation_loss)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training', 'validation'], loc='upper right')
plt.show()
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
%matplotlib inline

class_probabilities = model.predict(x_test)
predictions = np.argmax(class_probabilities, axis=1)
true_labels = np.argmax(y_test, axis=1)

# Plot the confusion matrix
cm = confusion_matrix(true_labels, predictions)
plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Reds)
plt.colorbar()
tick_marks = np.arange(2)
plt.xticks(tick_marks, [0, 1])
plt.yticks(tick_marks, [0, 1])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
