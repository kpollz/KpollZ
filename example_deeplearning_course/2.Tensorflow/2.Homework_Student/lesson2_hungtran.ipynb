```python
# 9. Từ bộ dữ liệu Income-Classification hãy chuẩn hóa dữ liệu và phân chia tập train/test theo tỷ lệ 80/20.
from google.colab import drive
import os

drive.mount('/content/drive')
os.chdir('/content/drive/MyDrive/Kaggle')
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('train.csv', header = 0)
# drop ID column since it is not relevant to train model
data = data.drop(['ID'], axis = 1)
# train['passive_income'] = train['capital_gain'] - train['capital_loss']
data.tail()
```

```python
data.info()
```

```python
data.describe()
```

```python
data = data.replace('?', np.nan)
```

```python
# fillna with mode
data = data.apply(lambda x: x.fillna(x.value_counts().index[0]))
data.info()
```

```python
for item in data.columns:
  print(item)
  print(data[item].value_counts())
```

```python
# One-hot encoder for category attributes
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[feature_to_encode])
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return(res)

features_to_encode = ['work_type', 'education', 'marital_state', 'job', 'status', 'ethnicity', 'sex', 'nationality']
for feature in features_to_encode:
    data = encode_and_bind(data, feature)

data.info()
```

```python
train, test = train_test_split(data, test_size = 0.2, stratify = data['target_income'], random_state = 0)
print(train.shape, test.shape)
```

```python
X_train = train.copy()
y_train = X_train.pop('target_income')
X_test = test.copy()
y_test = X_test.pop('target_income')
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## Câu 10

```python
# 10. Xây dựng một mạng deep-neural-network để huấn luyện mô hình trên dữ liệu train và đánh giá mô hình trên dữ liệu test.
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils
from tensorflow.keras import optimizers
```

```python
# Set data types for float features
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# Set data types for categorical labels
y_train = utils.to_categorical(y_train)
y_test = utils.to_categorical(y_test)
print('Ready...')
```

```python
# Define a classifier network
hl = 10 # Number of hidden layer nodes
target_income_classes = 2
features = 105

model = Sequential()
model.add(Dense(hl, input_dim=features, activation='relu'))
model.add(Dense(hl, input_dim=hl, activation='relu'))
model.add(Dense(target_income_classes, input_dim=hl, activation='sigmoid'))

print(model.summary())
```

```python
#hyper-parameters for optimizer
learning_rate = 0.0001
opt = optimizers.Adam(learning_rate=learning_rate)

model.compile(loss='binary_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])

# Train the model over 50 epochs using 10-observation batches and using the test holdout dataset for validation
num_epochs = 100
history = model.fit(X_train, y_train, epochs=num_epochs, batch_size=128, validation_data=(X_test, y_test))
```

The accuracy between train and test is quite the same => There is no overfitting

```python
# Visualization of the results
%matplotlib inline
from matplotlib import pyplot as plt

epoch_nums = range(1,num_epochs+1)
training_loss = history.history["loss"]
validation_loss = history.history["val_loss"]
plt.plot(epoch_nums, training_loss)
plt.plot(epoch_nums, validation_loss)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training', 'validation'], loc='upper right')
plt.show()
```

```python
# Tensorflow doesn't have a built-in confusion matrix metric, so we'll use SciKit-Learn
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
%matplotlib inline

target_income_classes = ['<50K', '>50K']

class_probabilities = model.predict(X_test)
predictions = np.argmax(class_probabilities, axis=1)
true_labels = np.argmax(y_test, axis=1)

# Plot the confusion matrix
cm = confusion_matrix(true_labels, predictions)
plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
plt.colorbar()
tick_marks = np.arange(len(target_income_classes))
plt.xticks(tick_marks, target_income_classes, rotation=85)
plt.yticks(tick_marks, target_income_classes)
plt.xlabel("Predicted Classification")
plt.ylabel("Actual Classification")
plt.show()
```
