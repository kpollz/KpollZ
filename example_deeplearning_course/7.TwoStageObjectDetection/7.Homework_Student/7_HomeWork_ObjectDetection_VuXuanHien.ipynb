#6) Xây dựng hàm tính toán IoU giữa hai bounding box.

```python
import numpy as np
```

```python
def IOU_scores(test, pred):
    # coordinate of intersection part
    x1 = max(test[0], pred[0])
    y1 = max(test[1], pred[1])
    x2 = min(test[0]+test[2], pred[0]+pred[2])
    y2 = min(test[1]+test[3], pred[1]+pred[3])

    # area of intersection
    intersec = (x2 - x1) * (y2 - y1)
    
    # area of 2 box
    test_area = test[2] * test[3]
    pred_area = pred[2] * pred[3]
    
    return intersec/(test_area + pred_area - intersec)
    
```

```python
test = np.array([20, 25, 30, 40])
pred = np.array([25, 30, 30, 45])

iou = IOU_scores(test, pred)
print('IOU: {:.2f}'.format(iou))
```

# 7)Hãy thực hiện vẽ đường cong Precision Recall curve và xây dựng thuật toán tính giá trị AP.

```python
import matplotlib.pyplot as plt
```

```python
box = np.array([[10, 10, 50, 20], [15, 15, 55, 25], [50, 20, 20, 25], [20, 50, 20, 50], [30, 20, 10, 10]])
ground_truth = np.array([[10, 10, 45, 25], [45, 20, 20, 25], [30, 25, 10, 10]])
```

```python
def collect_scores(box, ground_truth):
    TP = 0
    iter = 0
    scores = []
    precision = []
    recall = []

    for i in ground_truth:
      for j in box:
        scores.append(IOU_scores(i,j))
    scores.sort(reverse=True)

    for i in scores:
      iter += 1
      if i > 0.5:
        TP += 1

      precision.append(TP / iter)
      recall.append(TP / len(ground_truth))
    
    return precision, recall
```

```python
precision, recall = collect_scores(box, ground_truth)
```

```python
plt.figure(figsize=(8,6))
fig, ax = plt.subplots()
ax.plot(recall, precision,'o-')
plt.title('Precision Recall Curve')
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.show()
```

```python
def compute_AP(precision, recall):
    precision = np.concatenate([[0.0], precision, [0.0]])
    for i in range(len(precision) - 1, 0, -1):
        precision[i - 1] = np.maximum(precision[i - 1], precision[i])

    # find the index where the value changes
    recall = np.concatenate([[0.0], recall, [1.0]])
    changing_points = np.where(recall[1:] != recall[:-1])[0]

    # compute under curve area
    areas = (recall[changing_points + 1] - recall[changing_points]) * precision[changing_points + 1]
    return areas.sum() 
```

```python
print('AP:', compute_AP(precision,recall))
```

#9) Từ bộ dữ liệu Vinbigdata-chest-xray bao gồm các tọa độ bounding box của 14 nhãn của các loại bệnh khác nhau được chuẩn đoán trên ảnh chụp chest X-ray. Thực hiện biến đổi file dữ liệu train.csv sang định dạng COCO json format. Chi tiết mô tả về dữ liệu bạn đọc ở tab data link trên.

```python
import json
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/TwDS/data/train.csv')
df.head()
```

```python
def get_image(df):
  images = {}
  images['file_name'] = df['image_id'] + '.dcom'
  images['id'] = df['image_id']  
  return images
```

```python
def get_annotations(df):
  annotations = {}
  annotations['iscrowd'] = 0
  annotations['area'] = (df['y_max'] - df['y_min']) * (df['x_max'] - df['x_min']) 
  annotations['bbox'] = [df['x_min'], df['y_min'], df['x_max'] - df['x_min'], df['y_max'] - df['y_min']]
  annotations['category_id'] = df['class_id'] + 1
  return annotations
```

```python
images = []
for index, row in df.iterrows():
  images.append(get_image(row))

annotations = []
for index, row in df.iterrows():
  anno = get_annotations(row)
  anno['id'] = index
  annotations.append(anno)
```

```python
data = {}
data['images'] = images
data['annotations'] = annotations
json.dump(data, open('/content/drive/MyDrive/TwDS/data/data_coco.json', 'w'))
```
