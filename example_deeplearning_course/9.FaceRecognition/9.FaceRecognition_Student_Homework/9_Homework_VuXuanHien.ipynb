6) In this paper, they propose an Additive Angular Margin Loss (ArcFace) to further improve the discriminative power of the face recognition model and to stabilise the training process. 

In theory, they utilise the arc-cosine function to calculate the angle between the current feature and the target weight. Afterwards, add an additive angular margin to the target angle, and they get the target logit back again by the cosine function. Then, they re-scale all logits by a fixed feature norm, and the subsequent steps are exactly the same as in the softmax loss.

To compare ArcFacr with other method, despite the numerical similarity between ArcFace and previous works, the proposed additive angular margin has a better geometric attribute as the angular margin has the exact correspondence to the geodesic distance. They compare the decision boundaries under the binary classification case. The proposed ArcFace has a constant linear angular margin throughout the whole interval. By contrast, SphereFace and CosFace only have a nonlinear angular margin.

By training on datasets like LFW, CALFW, CPLFW, the ArcFace technique perform a very good result compares to others (up to 98%, 99%).

# Few-shot learning

```python
import torch 
import torch.nn as nn
import torchvision.transforms as T
from torch.utils.data import DataLoader
from torchvision.datasets import ImageFolder
import torch.nn.functional as F
```

```python
transform = T.Compose([
                       T.ToTensor(),
                       T.Resize((224,224)),
                       T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
dataset = ImageFolder(folder_name, transform=transform)
classes = dataset.classes
```

```python
data_loader = DataLoader(dataset=dataset, batch_size=32, shuffle=False, num_workers=2)
```

```python
class VGG_16(nn.Module):
    def __init__(self):
        super().__init__()
        self.block_size = [2, 2, 3, 3, 3]
        self.conv1_1 = nn.Conv2d(3, 64, 3, stride=1, padding=1)
        self.conv1_2 = nn.Conv2d(64, 64, 3, stride=1, padding=1)
        self.conv2_1 = nn.Conv2d(64, 128, 3, stride=1, padding=1)
        self.conv2_2 = nn.Conv2d(128, 128, 3, stride=1, padding=1)
        self.conv3_1 = nn.Conv2d(128, 256, 3, stride=1, padding=1)
        self.conv3_2 = nn.Conv2d(256, 256, 3, stride=1, padding=1)
        self.conv3_3 = nn.Conv2d(256, 256, 3, stride=1, padding=1)
        self.conv4_1 = nn.Conv2d(256, 512, 3, stride=1, padding=1)
        self.conv4_2 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        self.conv4_3 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        self.conv5_1 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        self.conv5_2 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        self.conv5_3 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        self.fc6 = nn.Linear(512 * 7 * 7, 4096)
        self.fc7 = nn.Linear(4096, 4096)

    def forward(self, x):
        """ Pytorch forward
        Args:
            x: input image (224x224)
        Returns: class logits
        """
        x = F.relu(self.conv1_1(x))
        x = F.relu(self.conv1_2(x))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv2_1(x))
        x = F.relu(self.conv2_2(x))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv3_1(x))
        x = F.relu(self.conv3_2(x))
        x = F.relu(self.conv3_3(x))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv4_1(x))
        x = F.relu(self.conv4_2(x))
        x = F.relu(self.conv4_3(x))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv5_1(x))
        x = F.relu(self.conv5_2(x))
        x = F.relu(self.conv5_3(x))
        x = F.max_pool2d(x, 2, 2)
        x = x.view(x.size(0), -1)
        # x = F.relu(self.fc6(x))
        # x = F.dropout(x, 0.5, self.training)
        # x = self.fc7(x)
        return x
```

```python
loss_f = nn.L1Loss()

def preprocess_image(img):
    img = img.resize((224, 224), Image.ANTIALIAS)
    try:
        img = np.array([img.getdata(0),img.getdata(1),img.getdata(2)])
    except:
        img = np.array([img.getdata(0),img.getdata(0),img.getdata(0)])
        
    img = torch.tensor(img.reshape(3,224,224)).float().to(device)
    return img

def predict(im1, im2):
    model.eval()
    with torch.no_grad():
        output1 = model(preprocess_image(im1).unsqueeze(0))
        output2 = model(preprocess_image(im2).unsqueeze(0))
    return loss_f(output1,output2)
```

# Triplet loss

```python
from PIL import Image
from glob import glob
import tensorflow as tf
import matplotlib.pyplot as plt
import tensorflow.keras.backend as K
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten, Dropout
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
```

```python
batch_size=32
IMG_SHAPE=(299, 299, 3)
IMG_SIZE = (299, 299)
```

```python
filename = os.listdir(folder_name)
label_dict = {K: V for V, K in enumerate(filename)}

def load_data(path=folder_name):
    
    x, y = [], []
    filepaths = glob(path+'/*')
    for file in filepaths:
        name = os.path.basename(file)
        x += glob(file+'/*')
        y += [label_dict[name]]*len(os.listdir(file))
    
    index = np.random.permutation(len(x))
    x, y = np.array(x)[index], np.array(y)[index]
    X_train, X_test = x[:13000], x[13000:]
    y_train, y_test = y[:13000], y[13000:]
    
    return (X_train, y_train), (X_test, y_test)

(X_train, y_train), (X_test, y_test) = load_data()
```

```python
def Image_Preprocessing(path):
    
    image = Image.open(path).resize(IMG_SIZE)
    image = np.asarray(image)
    '''
    image = np.reshape(image, IMG_SHAPE)
    image = np.expand_dims(image, axis=0)
    image = tf.cast(image, tf.float32)
    image = image/255.0
    '''
    image = preprocess_input(image, data_format='channels_last')
    return image

def get_image(label, test=False):
    
    X, y = (X_test, y_test) if test else (X_train, y_train)
    
    idx = [i for i in range(len(y)) if y[i]==label]
    index = np.random.choice(idx)
    img_path = X[index]
    image = Image_Preprocessing(img_path)
    return image

def get_triplet(test=False):
    
    n = a = np.random.randint(105)
    while n == a:
        # keep searching randomly!
        n = np.random.randint(105)
    a, p = get_image(a, test), get_image(a, test)
    n = get_image(n, test)
    return a, p, n
```

```python
def generate_triplets(test=False):
    """Generate an un-ending stream (ie a generator) of triplets for
    training or test."""
    while True:
        list_a = []
        list_p = []
        list_n = []

        for i in range(batch_size):
            a, p, n = get_triplet(test)
            list_a.append(a)
            list_p.append(p)
            list_n.append(n)
            
        A = np.array(list_a, dtype='float32')
        P = np.array(list_p, dtype='float32')
        N = np.array(list_n, dtype='float32')
        # a "dummy" label which will come in to our identity loss
        # function below as y_true. We'll ignore it.
        label = np.ones(batch_size)
        yield [A, P, N], label

train_generator = generate_triplets()
test_generator = generate_triplets(test=True)
```

```python
def identity_loss(y_true, y_pred):
    return K.mean(y_pred)


def triplet_loss(y_pred, alpha=0.6):
    
    anchor, positive, negative = y_pred[0], y_pred[1], y_pred[2]
    pos_dist = euclidean_distance([anchor, positive])
    neg_dist = euclidean_distance([anchor, negative])
    basic_loss = pos_dist - neg_dist + alpha
    loss = K.sum(K.maximum(basic_loss, 0), axis=-1, keepdims=True)
    
    return loss


def euclidean_distance(vectors):
    
    #unpack the vectors into separate lists
    (featsA, featsB) = vectors
    
    #computing the sum of squared distances between the vectors
    sumSquared = K.sum(K.square(featsA - featsB), axis=-1, keepdims=True)
    
    #returning the euclidean distance
    return K.sqrt(K.maximum(sumSquared, K.epsilon()))
```

```python
def encoding_model():
    
    model = InceptionV3()
    
    for i in range(18):
        model.layers[i].trainable = False
    
    x = model.layers[-1].output
    x = Flatten()(x)
    x = Dropout(0.2)(x)
    x = Dense(1024, activation='relu')(x)
    x = Lambda(lambda x: K.l2_normalize(x, axis=1))(x)
    
    encoder = Model(inputs=model.input, outputs=x, name='EncodingModel')
    
    return encoder

def siamese_model():
    
    anchor = Input(IMG_SHAPE)
    posImage = Input(IMG_SHAPE)
    negImage = Input(IMG_SHAPE)
    
    encoder = encoding_model()
    
    anchorEncoded = encoder(anchor)
    posImageEncoded = encoder(posImage)
    negImageEncoded = encoder(negImage)
    
    loss = Lambda(triplet_loss)([anchorEncoded, posImageEncoded, negImageEncoded])
    
    siamese_network = Model(inputs=[anchor, posImage, negImage], outputs=loss, name='siamese_network')
    
    return siamese_network

model = siamese_model()
model.summary()
```

```python
lr = 1e-4
model.compile(optimizer=Adam(lr), loss=identity_loss)

model.fit(train_generator, validation_data=test_generator,verbose=1, epochs=2, steps_per_epoch=512,
          validation_steps=256)
```
