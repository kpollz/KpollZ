## 6. Tổng kết ý chính trong paper
* The paper proposes an Additive Angular Margin Loss (ArcFace) to obtain highly discriminative features for face recognition and to stabilise the training
process.
* To interpret this, the writers have compared ArcFace with SphereFace, CosFace and other losses. They also did some experiments on various datasets such as LFW, YTF, CFP-FP, AgeDB-30, IJB-B,C...to check the improvement from different settings. 
* Finally, they have demonstrated that ArcFace consistently outperforms the state of the art through the most
comprehensive experiments

```python
import os
main_folder = '/content/drive/MyDrive/celebrity_images/105_classes_pins_dataset'
print('The number of celebrities:',len(os.listdir(main_folder)))
```

```python
# the average number of images per person
import pandas as pd
people = []
num_imgs = []
for person_name in os.listdir(main_folder):
  people.append(person_name)
  count = len(os.listdir(os.path.join(main_folder,person_name)))
  num_imgs.append(count)
df = pd.DataFrame({'Person': people, 'Imgs_num': num_imgs}).sort_values('Imgs_num')
df
```

```python
df.describe()
```

```python
#Visualization:
import matplotlib.pyplot as plt
plt.figure(figsize=(8,20))
plt.barh(df['Person'], df['Imgs_num'])
plt.ylabel('Person')
plt.xlabel('Number of images')
plt.title('Number of images per person')
```

```python
from retinaface import RetinaFace
```

```python
img_path = '/content/drive/MyDrive/celebrity_images/105_classes_pins_dataset/pins_scarlett johansson/scarlett johansson241_4018.jpg'
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,6))
img = plt.imread(img_path)
plt.imshow(img)
```

```python
obj = RetinaFace.detect_faces(img_path)
obj
```

```python
# detection
import cv2
facial_area = obj['face_1']['facial_area']
cv2.rectangle(img, (facial_area[2], facial_area[3]), (facial_area[0], facial_area[1]), (225,0,0), 1)
plt.figure(figsize=(6,6))
plt.imshow(img)
```

```python
# crop and align image
faces = RetinaFace.extract_faces(img_path = img_path, align = True)
plt.imshow(faces[0])
plt.figure(figsize=(6,6))
plt.show()
```
