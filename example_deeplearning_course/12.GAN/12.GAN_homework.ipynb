# I. Lý thuyết

1) Lớp mô hình generative model là gì?

A. Là lớp mô hình học cách sinh ra dữ liệu từ phân phối của tập huấn luyện.

B. Là lớp mô hình nhằm dự báo giá trị từ dữ liệu đầu vào.

C. Là lớp mô hình phân cụm dữ liệu.

D. Là lớp mô hình phân loại nhãn dựa vào dữ liệu đầu vào.


2) Sự khác biệt giữa các lớp mô hình generative model sử dụng explicit density và implicit density là gì?

A. Explicit density sẽ tạo ra một mô hình sinh mẫu dữ liệu mà không cần xác định phân phối xác suất của dữ liệu. Implicit density sẽ tìm ra phân phối xác suất dựa trên ước lượng hàm mật độ xác suất.

B. Explicit density sẽ tìm ra phân phối xác suất dựa trên ước lượng hàm mật độ xác suất. Phương pháp implicit density sẽ tạo ra một mô hình sinh mẫu dữ liệu mà không cần xác định phân phối xác suất của dữ liệu.

C. Explicit density sẽ tìm ra xấp xỉ của hàm mật độ xác suất trong khi Implicit density sử dụng một mạng nơ ron để sinh mẫu.

D. Explicit density không tìm ra trực tiếp phân phối xác suất vì đây là phương pháp ẩn còn Implicit density sẽ tìm ra trực tiếp phân phối xác suất nếu đây là phương pháp hiện.


3) Variational Autoencoder là mô hình như thế nào ? Có thể chọn nhiều hơn 1 đáp án.

A. Mô hình được xây dựng dựa trên tối ưu hóa trên chính xác hàm likelihood.

B. Mô hình cũng tìm cách tối đa hóa likelihood function nhưng giá trị của likelihood function là bất trị (intractable) và chúng ta cần huấn luyện mô hình dựa trên tối ưu hóa lower bound của hàm log likelihood. 

C. Kiến trúc gồm hai phases là Encoder và Decoder trong đó: Cả Encoder và Decoder đều sinh mẫu dựa trên ước lượng tham số của phân phối xác suất bao gồm trung bình và ma trận hiệp phương sai.

D. Là một generative model thuộc họ explicit density nhưng chỉ có thể ước lượng được giá trị xấp sỉ của hàm density.


4) Nguyên lý hoạt động chính của GAN là gì?

A. Là một trò chơi hai người: Generator cố gắng đánh lừa Discriminator bằng cách sinh ra ảnh giả giống với thật. Discriminator cố gắng phân biệt giữa ảnh giả và ảnh thật.

B. Là một generative model sinh mẫu dựa trên phân phối xác suất

C. Là một trò chơi hai người: Discriminator cố gắng đánh lừa Generator bằng cách sinh ra ảnh giả giống với thật. Discriminator cố gắng phân biệt giữa ảnh giả và ảnh thật.

D. Một kiến trúc GAN sẽ gồm hai mô hình: Generator có tác dụng sinh ảnh giả, Discriminator có tác dụng phân loại nhãn của object chứa bên trong ảnh.
