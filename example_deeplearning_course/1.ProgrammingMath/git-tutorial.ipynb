# Git first commit

Next we need to add a new file README.md into the remote repository name `git-tutorial`.

* Step 1: Create REAME.md file
* Step 2: Commit the change by message
* Step 3: Create main branch name main
* Step 4: Attach remote url into origin
* Step 5: Push the change in stage into remote repo

```
echo "# git-tutorial" >> README.md
git add README.md # Add file README.md into stage
git commit -m "first commit" # commit the change by message
git branch -M main # create main branch name main
git remote add origin https://github.com/towarddatascience/git-tutorial.git # attach remote url into origin
git push -u origin main # push the change into remote repo
```

# Git create a new branch
